```python
import numpy as np
```

```python
rng=np.random.default_rng(1000)
```

```python
n=20
x=rng.random(n+1)*100
```

```python
y=rng.random(n+1)*200
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
ax,fig=plt.subplots(figsize=(10,10))
ax=plt.scatter(x[1:],y[1:],color='green',s=100)
ax=plt.scatter(x[0],y[0],color='red',s=100,marker='s')
```
